```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
dataset=pd.read_csv("Original_Dataset.csv")
dataset
```

```python
symptoms=[]
## this loop is for creating new column called Symptoms that contain each row's values 
for i in range(len(dataset)):
    value=dataset.iloc[i].values.tolist()## values of each row
    if 0 in value:
         symptoms.append(value[1:value.index(0)]) ## if value is 0 only append data in row that is not 0
    else:
        symptoms.append(value[1:])
        
dataset["Symptoms"]=symptoms
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

column_values=dataset[["Symptom_1","Symptom_2","Symptom_3","Symptom_4","Symptom_5","Symptom_6","Symptom_7","Symptom_8","Symptom_9",
"Symptom_10","Symptom_11","Symptom_12","Symptom_13","Symptom_14","Symptom_15","Symptom_16","Symptom_17"]].values.ravel()
symps=pd.unique(column_values).tolist() 
symps=[i for i in symps if str(i)!="nan"] #
```

```python

new_data2=pd.DataFrame(columns=dataset.columns,index=dataset.index)

x2=new_data2.drop(["Disease","Symptoms"],axis=1)
columns=x2.columns.tolist()
columns
for i in columns:
       x2[i]=dataset[i].apply(lambda x2: symps.index(x2)+1 if x2 in symps else 0)

encode=LabelEncoder()
y2=dataset["Disease"]
#y2=encode.fit_transform(y2)
df=pd.DataFrame(y2)
y_encoded = pd.get_dummies(df, dtype=int)
y_encoded
```

## Hyper Parameters

 4. optimizer 
 5. Tune batch size and number of epochs

```python
# define baseline model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

def baseline_model(neurons, hidden_layers):
    # create model
    model = Sequential()
    model.add(Dense(neurons, input_shape=(17,), activation='relu'))
    for i in range(hidden_layers):
        model.add(Dense(neurons, activation='relu'))      
    model.add(Dense(41, activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python

#estimator = KerasClassifier(build_fn=baseline_model, epochs=80, batch_size=5, verbose=1)
pipeline=Pipeline([('scaler',StandardScaler()),('model',KerasClassifier(build_fn=baseline_model,epochs=50))])
Hyper_Peramaters={
'model__batch_size' : [10,20],
'model__model__neurons': [80,100],
'model__model__hidden_layers':[1,2]
    
}
grid = GridSearchCV(pipeline,Hyper_Peramaters ,cv=5)
grid.fit(x2, y_encoded)
print("Best peramters combination is ")
print(grid.best_params_)
print("Best Estimator ")
print(grid.best_estimator_)
print("Best cross-validation score:  ")
print(grid.best_score_)


```

```python
final_model=grid.best_estimator_
final_model
```

```python
from sklearn.model_selection import train_test_split,cross_val_score,KFold,ShuffleSplit,LeavePOut


shuffle_split = ShuffleSplit(n_splits=10, test_size=0.3, random_state=0)
k_folds = KFold(n_splits = 10,shuffle=True)
scores = cross_val_score(final_model, x2, y_encoded, cv=10, scoring='accuracy')
scores_shuffle_split = cross_val_score(final_model, x2, y_encoded, cv=shuffle_split, scoring='accuracy')
scores_k_folds = cross_val_score(final_model, x2, y_encoded, cv=k_folds, scoring='accuracy')
# scores = cross_val_score(pipeline, x2, y2, cv=10, scoring='accuracy') 

print("scores cv=10")
print(scores.mean())

print("scores cv=shuffle_split")
print(scores_shuffle_split.mean())
print("scores cv=k_folds")
print(scores_k_folds.mean())
```

```python
X_train,X_test,y_train,y_test=train_test_split(x2,y_encoded,test_size=0.3)
final_model.fit(X_train,y_train)
y_pred=final_model.predict(X_test)
```
